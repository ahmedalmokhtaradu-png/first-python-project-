### How to remove items from a list
To remove an item from a list, you use the `remove()` command.
Syntax: `list_name.remove(item_to_remove)`

```python
the_grocery_list = ["tomato", "botato", "deodrant", "orange_juice"]
print(f"Current list: {the_grocery_list}")

# 1. Ask what to remove
item_to_remove = input("What are you checking out? ")

# 2. Check if it's in the list before trying to remove it
if item_to_remove in the_grocery_list:
    the_grocery_list.remove(item_to_remove)
    print(f"Success! '{item_to_remove}' has been checked out.")
else:
    print(f"Error: '{item_to_remove}' is not in your list!")

# 3. Show the updated list
print(f"Final list: {the_grocery_list}")
```
